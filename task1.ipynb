```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('world_population.CSV',  encoding='utf-8')
data.head()
```

```python
data_final = data.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
chosen_country = 'Belgium'


country_data = data_final[data_final['Country Name'] == chosen_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
